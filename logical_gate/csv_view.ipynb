### 指数関数でフィッティング

```python
def logical_error_list(p,code_distance_list):
    return 1/2 * (1-(1-2*p)**code_distance_list)

def dif_loss_fun(data_list,logical_error_list,p,code_distance_list):
    answer = 0
    #data_list = data_list.values
    for cd in code_distance_list:
        answer += cd * (1-2*p)**int((cd-3)/2) * (data_list[int((cd-3)/2)] - logical_error_list[int((cd-3)/2)])
    return answer

def fit(data_list,initial_p,code_distance_list,trials,alpha):
    # trials：試行回数
    # alpha：学習率
    p = initial_p # pの初期値
    data_list = np.array(data_list)
    # ロス関数
    #for i in round_list:
        #loss_fun = 1/2 * np.sum((data_list - logical_error_list(p,np.arange(1,round_list,1)))**2)
    for trial in range(trials):
        # pの更新
        p = p + alpha * dif_loss_fun(data_list,logical_error_list(p,code_distance_list),p,code_distance_list)
        loss_fun = 1/2 * np.sum((data_list - logical_error_list(p,code_distance_list))**2)
        #print(p, loss_fun)
    return p
```
