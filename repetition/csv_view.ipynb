## 普通にグラフを作成(誤り確率に沿って重み付け)

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import os

################ 要変更 #################
# ファイル名
file = "p=0.5% ,eta=1000, d=11, rep=10, # of trials=100000" + ".csv"
########################################

# FUjii-labのパス
path = "/Users/konohideki/Documents/Fujii-lab/repetition"

# csvファイルの読み込み
os.chdir(path)
data = pd.read_csv(file, index_col=0)

# データフレームの行名(リスト)を取得
ind = data.index

# データフレームの列名(リスト)を取得
code_distance_list = data.columns

##### グラフ作成 #####

plt.rcParams["xtick.direction"] = "in"     
plt.rcParams["ytick.direction"] = "in" 
fig, ax = plt.subplots()
a = np.array(code_distance_list)
a = [float(num) for num in a]
a = np.array(a)
b = np.array(data.iloc[0])
ax.plot(a[0:], b[0:]*100,marker='v',label="logical X error")
b = np.array(data.iloc[1])
ax.plot(a[0:], b[0:]*100,marker='v',label="logical Z error")
#ax.set_xticklabels(np.round(a,decimals= 2),rotation=45)
ax.set_xticks(a)
ax.set_xlabel("code distance", fontsize=13)
ax.set_ylabel("logical failure probability (%)", fontsize=13)
#ax.set_xlim(9,11)
#ax.set_ylim(12,17)
ax.spines["top"].set_linewidth(2)
ax.spines["left"].set_linewidth(2)
ax.spines["bottom"].set_linewidth(2)
ax.spines["right"].set_linewidth(2)
ax.tick_params(direction="in", width=2, length=4, labelsize=12)
#ax.set_title("X,p=(0.0001,0.0009,0.0001),d=(3,9,2),d(rep)=5,eta=1000,trials=10000", fontsize=14)
plt.legend()
plt.show()
```

```python
e = 1/2 * (1 - (1 - 2 * b)**(1/10))
print(e)
```

```python
p = 0.1
file1 = "p=" + str(p) + "% ,eta=1000, d=11, rep=10, # of trials=100000" + ".csv"
file2 = "p=" + str(p) + "% ,eta=1000, d=11, rep=20, # of trials=100000" + ".csv"
file3 = "p=" + str(p) + "% ,eta=1000, d=11, rep=30, # of trials=100000" + ".csv"
file4 = "p=" + str(p) + "% ,eta=1000, d=11, rep=40, # of trials=100000" + ".csv"
file5 = "p=" + str(p) + "% ,eta=1000, d=11, rep=50, # of trials=100000" + ".csv"

data1 = pd.read_csv(file1, index_col=0)
data2 = pd.read_csv(file2, index_col=0)
data3 = pd.read_csv(file3, index_col=0)
data4 = pd.read_csv(file4, index_col=0)
data5 = pd.read_csv(file5, index_col=0)

round = [10,20,30,40,50]
logical_error_rate = []
fig, ax = plt.subplots()
for i in range(5):
    if i == 0:
        continue
    logical_error_rate = []
    logical_error_rate.append(data1.iloc[1][i])
    logical_error_rate.append(data2.iloc[1][i])
    logical_error_rate.append(data3.iloc[1][i])
    logical_error_rate.append(data4.iloc[1][i])
    logical_error_rate.append(data5.iloc[1][i])
    logical_error_rate = np.array(logical_error_rate)
    ax.plot(round, logical_error_rate*100,marker='v',label="code distance = " + str(2*i+3))
ax.set_xticks(round)
ax.set_xlabel("round", fontsize=13)
ax.set_ylabel("logical failure probability (%)", fontsize=13)
#ax.set_xlim(9,11)
#ax.set_ylim(12,17)
ax.spines["top"].set_linewidth(2)
ax.spines["left"].set_linewidth(2)
ax.spines["bottom"].set_linewidth(2)
ax.spines["right"].set_linewidth(2)
ax.tick_params(direction="in", width=2, length=4, labelsize=12)
#ax.set_title("X,p=(0.0001,0.0009,0.0001),d=(3,9,2),d(rep)=5,eta=1000,trials=10000", fontsize=14)
plt.legend()
plt.show()
```
