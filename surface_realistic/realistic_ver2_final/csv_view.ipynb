```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import os

plt.rcParams["xtick.direction"] = "in"     
plt.rcParams["ytick.direction"] = "in" 
fig, ax = plt.subplots()
p_list = np.array([1e-03, 2e-03, 3e-03, 4e-03, 5e-03])
e_3 = np.array([6.221135056124972e-05, 0.0003055093520170069, 0.0006572927378995785, 0.0011410234133757842, 0.00174445284677083])
e_5 = np.array([9.24222576898638e-06, 5.407981073186522e-05, 0.00014841342587035088, 0.0003036979977304236, 0.0005246252224646754])
e_7 = np.array([4.028094312702551e-06, 1.9911523076237106e-05, 5.640787094903298e-05, 0.00012115535493900043, 0.00022020628078973608])
e_9 = np.array([2.930722943959408e-07, 3.6743365142721676e-06, 1.3665902871521332e-05, 3.6132132058039436e-05, 7.635393302886187e-05])
e_11 = np.array([1.0157284721615325e-07, 7.914019585422747e-07, 3.33900084574834e-06, 1.0865748319900874e-05, 2.519769256908906e-05])

a = np.array(p_list)
ax.plot(p_list, e_3,marker='v',label="d="+str(3))
ax.plot(p_list, e_5,marker='v',label="d="+str(5))
ax.plot(p_list, e_7,marker='v',label="d="+str(7))
ax.plot(p_list, e_9,marker='v',label="d="+str(9))
ax.plot(p_list, e_11,marker='v',label="d="+str(11))
ax.set_xticklabels(np.round(a,decimals= 1))
ax.set_xticks(np.round(a,decimals= 2))
ax.set_xlabel("physical error probability (%)", fontsize=13)
ax.set_ylabel("logical failure probability (%)", fontsize=13)
#ax.set_ylim(10**(-3))
#ax.set_xlim(10**(-10))
ax.spines["top"].set_linewidth(2)
ax.spines["left"].set_linewidth(2)
ax.spines["bottom"].set_linewidth(2)
ax.spines["right"].set_linewidth(2)
ax.tick_params(direction="in", width=2, length=4, labelsize=12)
#ax.set_title("pm=" + str(0) + "%, eta=" + str(1000) + ", repeat=" + str(1) + ", trials=" + str(10000), fontsize=14)
ax.set_xscale("log")
ax.set_yscale("log")
ax.grid()
plt.legend()
plt.show()
```
