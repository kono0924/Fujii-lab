## MultiZ

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import os
import matplotlib.ticker as ptick

################ 要変更 #################

# FUjii-labのパス
path = "/Users/konohideki/Documents/Fujii-lab/surface_realistic/realistic_multi_round/multiZ,d1=3,p=0.0001,eta=1000"

# ファイル名

file1 = "X error,p=0.0001,d2=(3,9,2),eta=1000,trials=10000" + ".csv"
file2 = "Z error,p=0.0001,d2=(3,9,2),eta=1000,trials=10000" + ".csv"
########################################

# csvファイルの読み込み
os.chdir(path)
data1 = pd.read_csv(file1, index_col=0)
data2 = pd.read_csv(file2, index_col=0)

# データフレームの行名(リスト)を取得
code_distance_list = data1.index

# データフレームの列名(リスト)を取得
round_list = data1.columns

##### グラフ作成 #####

a = np.array(round_list)
a = [float(num) for num in a]
a = np.array(a)

cm = plt.get_cmap("tab10")

fig, axes = plt.subplots(1,2,figsize=(18,6))
for i in range(len(code_distance_list)-1):
    b = np.array(data1.loc[code_distance_list[i]])
    c = np.array(data2.loc[code_distance_list[i]])
    axes[0].plot(a[0:]+1, b[0:],marker='x',label=r"$(X \mathrm{error})~ d_2=$"+str(code_distance_list[i]),color = cm(i))
    axes[0].plot(a[0:]+1, c[0:],marker='v',label=r"$(Z \mathrm{error})~ d_2=$"+str(code_distance_list[i]),color = cm(i))
    axes[1].plot(a[0:]+1, b[0:],marker='x',label=r"$(X \mathrm{error})~ d_2=$"+str(code_distance_list[i]),color = cm(i))
    axes[1].plot(a[0:]+1, c[0:],marker='v',label=r"$(Z \mathrm{error})~ d_2=$"+str(code_distance_list[i]),color = cm(i))
axes[0].set_xlabel("# of divitions", fontsize=13)
axes[0].set_ylabel("logical failure probability", fontsize=13)
axes[0].set_ylim(0,)
axes[0].yaxis.set_major_formatter(ptick.ScalarFormatter(useMathText=True)) 
axes[0].ticklabel_format(style="sci",  axis="y",scilimits=(0,0))
axes[1].set_xlabel("# of divitions", fontsize=13)
axes[1].set_ylabel("logical failure probability", fontsize=13)
# axes.spines["top"].set_linewidth(2)
# axes.spines["left"].set_linewidth(2)
# axes.spines["bottom"].set_linewidth(2)
# axes.spines["right"].set_linewidth(2)
axes[0].tick_params(direction="in", width=2, length=4, labelsize=12)
axes[1].set_yscale("log")
plt.legend()
plt.show()
```
