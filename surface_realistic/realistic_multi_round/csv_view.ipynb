## MultiZ

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import os
import matplotlib.ticker as ptick

################  #################

# FUjii-lab
path = "/Users/konohideki/Documents/Fujii-lab/surface_realistic/realistic_multi_round/multiX,d1=5,p=0.0001,eta=1000"

# 

file1 = "X error,p=0.0001,d2=(3,9,2),eta=1000,trials=200000" + ".csv"
file2 = "Z error,p=0.0001,d2=(3,9,2),eta=1000,trials=200000" + ".csv"
########################################

# csv
os.chdir(path)
data1 = pd.read_csv(file1, index_col=0)
data2 = pd.read_csv(file2, index_col=0)

# ()
code_distance_list = data1.index

# ()
round_list = data1.columns

#####  #####

a = np.array(round_list)
a = [float(num) for num in a]
a = np.array(a)

cm = plt.get_cmap("tab10")

fig, axes = plt.subplots(1,2,figsize=(18,6))
for i in range(len(code_distance_list)):
    b = np.array(data1.loc[code_distance_list[i]])
    c = np.array(data2.loc[code_distance_list[i]])
    axes[0].plot(a[0:], b[0:],marker='v',label="X,d2="+str(code_distance_list[i]),color = cm(i))
    axes[0].plot(a[0:], c[0:],marker='x',label="Z,d2="+str(code_distance_list[i]),color = cm(i))
    axes[1].plot(a[0:], b[0:],marker='v',label="X,d2="+str(code_distance_list[i]),color = cm(i))
    axes[1].plot(a[0:], c[0:],marker='x',label="Z,d2="+str(code_distance_list[i]),color = cm(i))
axes[0].set_xlabel("div", fontsize=13)
axes[0].set_ylabel("logical failure probability", fontsize=13)
axes[0].set_ylim(0,)
axes[0].yaxis.set_major_formatter(ptick.ScalarFormatter(useMathText=True)) 
axes[0].ticklabel_format(style="sci",  axis="y",scilimits=(0,0))
axes[1].set_xlabel("div", fontsize=13)
axes[1].set_ylabel("logical failure probability", fontsize=13)
# axes.spines["top"].set_linewidth(2)
# axes.spines["left"].set_linewidth(2)
# axes.spines["bottom"].set_linewidth(2)
# axes.spines["right"].set_linewidth(2)
axes[0].tick_params(direction="in", width=2, length=4, labelsize=12)
axes[0].set_title("X error", fontsize=14)
axes[1].tick_params(direction="in", width=2, length=4, labelsize=12)
axes[1].set_title("X error(log-log)", fontsize=14)
axes[1].set_xscale("log")
axes[1].set_yscale("log")
plt.legend()
plt.show()
```
